```python
#--------------------make guess number game set seceret number to 5------------------
seceret_number = 5
attempts = 0
while True:
  guess = int(input("guess a number:"))
  attempts += 1
  if attempts > 5:
    print("Your attempts are over. You lose!")
    break
  if guess == seceret_number:
    print("Congratulations! You attempted to guess the number.", attempts)
    break
  elif guess < seceret_number:
    print("Too Low! Try again.")
  else:
    print("Too high! Try again.")
```

```python
#---------------Atm Pin Checker Using attempts using while loop------------------
pin_code = 1234
attempts = 3
while True:
    enter_pin = int(input("enter your pin code:"))
    if enter_pin == pin_code:
        print("pin code is correct")
        break
    else:
        attempts -= 1
        print("pin code is incorrect. Attempts left:", attempts)
        if attempts == 0:
            print("your card is blocked.")
            break
        print("try again")
```
