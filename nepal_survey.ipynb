step 6a Fill with mean
use mean value with normal distribution like 2.90, 2.40

```python
#-------------------check specific columns missing values-------------
print("Missing values in 'household_size':", df_clean["household_size"].isnull().sum()) #checking missing values in household size column

#---------calculate the mean average------------------
mean_value = df_clean["household_size"].mean()#calculating the mean average of household size
print("Mean average of household size:", round(mean_value, 1)) #rounding to 1 decimal place

#Fill missing values with mean average
df_clean["household_size"] = df_clean["household_size"].fillna(mean_value) #filling missing values in household size column with mean average
print("missing values in household_size after filling :", df_clean["household_size"].isnull().sum()) #checking missing values in household size column after filling with mean average
```

```python
mean= df_clean["land_ownership_ropani"].mean() #filling missing values in land ownership ropani column with mean average
mid  =df_clean["land_ownership_ropani"].median() #filling missing values in land ownership ropani column with median average
print(mean, mid)
```

6b - fill with median
use median for numeric columns that have outliners like salary,land ownership-----

```python
#---------------fill monthly_income column with median average----------------
print("Missing values in 'Monthly income':", df_clean["monthly_income_nrs"].isnull().sum()) #checking missing values in monthly income column

median_income = df_clean["monthly_income_nrs"].median() #calculating the median average of monthly income
print("Median monthly income (NPR):", median_income)  #rounding to 2 decimal places

df_clean["monthly_income_nrs"] = df_clean["monthly_income_nrs"].fillna(median_income) #filling missing values in monthly income column with median average

print("Missing in monthly income after filling:", df_clean["monthly_income_nrs"].isnull().sum()) #checking missing values in monthly income column after filling with median average
```

```python
print("Missing values in 'land_ownership_ropani' before filling:", df_clean["land_ownership_ropani"].isnull().sum()) #checking missing values in land ownership ropani column before filling

median_land = df_clean["land_ownership_ropani"].median() #calculating the median average of land ownership ropani
print("Median land ownership (ropani):", round(median_land ,2)) #rounding to 2 decimal places

df_clean["land_ownership_ropani"] = df_clean["land_ownership_ropani"].fillna(median_land) #filling missing values in land ownership ropani column with median average
print("missing value in land_ownership_ropani after filling:", df_clean["land_ownership_ropani"].isnull().sum()) #checking missing values in land ownership ropani column after filling with median average
```

```python
print("Missing value in 'annual_remmitance_nrs' before filling:", df_clean["annual_remittance_nrs"].isnull().sum()) #checking missing values in annual remittance column before filling
median_remittance = df_clean["annual_remittance_nrs"].median() #calculating the median average of annual remittance
print("Median annual remittance (NPR):", round(median_remittance, 2)) #rounding to 2 decimal places
df_clean["annual_remittance_nrs"] = df_clean["annual_remittance_nrs"].fillna(median_remittance) #filling missing values in annual remittance column with median average
print("Missing values in annual_remittance_nrs after filling:", df_clean["annual_remittance_nrs"].isnull().sum()) #checking missing values in annual remittance column after filling with median average
```

6c mode() for text or category columns (most common value)
frequency of distribution or more repeated 
Example:education_level,occupation

```python
df_clean["education_level"].value_counts() #checking the value counts of education level column to find the mode value
```

```python
#-------------------------Fill education level with mode------------------
print("Missing values in 'education_level' before filling:", df_clean["education_level"].isnull().sum()) #checking missing values in education level column before filling
mode_education = df_clean["education_level"].mode()[0] #calculating the mode
print("Mode of education level:", mode_education) #printing the mode value of education level column
df_clean["education_level"] = df_clean["education_level"].fillna(mode_education) #filling missing values in education level column with mode
print("Missing values in education level after filling:", df_clean["education_level"].isnull().sum()) #checking missing values in education level column after filling with mode
```

```python
df_clean["occupation"].value_counts() #checking the value counts of occupation column to find the mode value
```

```python
print("Missing values in 'occupation' before filling:", df_clean["occupation"].isnull().sum()) #checking missing values in occupation column before filling
mode_occupation = df_clean["occupation"].mode()[0] #calculating the mode value
print("Mode of occupation:", mode_occupation) #printing the mode value of occupation column
df_clean["occupation"] = df_clean["occupation"].fillna(mode_occupation) #filling missing values in occupation column with mode
print("Missing values in occupation after filling:", df_clean["occupation"].isnull().sum()) #checking missing values in occupation column after filling with mode
```

```python
#------------6d forward fill method for toilet facility column----------------
df_clean["health_insurance"] =df_clean["health_insurance"].ffill() #filling missing values in health insurance column with forward fill method
print("Missing value in health_insurance after filling:", df_clean["health_insurance"].isnull().sum()) #checking missing values in health insurance column after filling with forward fill method
```
