```python
ct_age = pd.crosstab(heart["Age_Group"], heart["Heart_Attack_Incidence"])
plt.figure(figsize=(15,5))
sns.heatmap(ct_age, annot= True, fmt="d", cmap="YlGnBu")
plt.title("Cross Tabulation Between Age_Group and Heart Attack Incidence")
plt.xlabel("Heart Attack Incidence")
plt.ylabel("Age Group")
plt.show()
```
