```python
#---------step 1: check for missing values-------------------
missing_count = df .isnull().sum()   #----------------missing count check ----------------------
print("Missing coun:")
print(missing_count)
```

```python
#------------step 2: check for missing values in percentage-------------------
missing_percent = (df.isnull().sum() / len(df)) * 100   #---------------missing count in percentage-----------------------------
print("Mising count in percent")
print(missing_percent)
```

```python
#-----------------show only columns that have atleast one missing values-------------------
print(missing_percent.round(2)) #-----------------rounding off the percentage to 2 decimal places-------------------
```

```python
#------step 3: show only rows that have atleast one missing values-------------------
#--------show only columns that have atleast one missing values-------------------
columns_with_mising = df.columns[df.isnull().any()].tolist() #-----------------columns with missing values-------------------
print("columns with missing values", len(columns_with_mising))
print(columns_with_mising)
```

```python
#----------------------show only columns that have atleast one missing values-------------------
#---------show only rows that have atleast one missing values-------------------
rows_with_missing = df[df.isnull().any(axis=1)] #-----------------rows with missing values-------------------
print("rows with missing values", len(rows_with_missing))
print(" total rows with missing values:", len(rows_with_missing))
rows_with_missing
```

```python
#----------------------step 4:visualization of missing values using bar plot-------------------
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (14, 6)) #------------------visualization of missing values using bar plot-------------------
plt.title("Missing Value Count - Dark Means missing") #------------------title of the plot-------------------
plt.imshow(df.isnull(), aspect = "auto", cmap ="Greys", interpolation = "none") #------------------visualization of missing values using bar plot-------------------
plt.xticks(range(len(df.columns)), df.columns, rotation =45, ha = "right") #------------------x-axis labels-------------------
plt.yticks(range(len(df)), df["StudentID"]) #------------------y-axis labels-------------------
plt.xlabel("Columns_Name") #------------------x-axis label-------------------
plt.ylabel("StudentID") #------------------y-axis label-------------------
plt.tight_layout() #------------------adjust layout to prevent overlap-------------------
plt.show() #------------------display the plot-------------------
```
