```python
#-----------Guess the number game-------------
secret = 7
attempts = 0

while True:
    guess = int(input("guess the number but you have only 3 chances:"))
    attempts += 1
    if attempts > 3:
        print("your limit is over: game over.")
        break
    if guess == secret:
        print("correct Attempts:", attempts)
        break
    elif guess < secret:
        print("too low")
    else:
        print("too high")
```
