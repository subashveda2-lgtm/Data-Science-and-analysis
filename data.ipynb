```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\ssita\pythonclass\data.csv")
print("Shape of the dataset", df.shape)
df
```

```python
df.info()
```

```python
missing_values = df.isnull().sum()
print("Missing value count:")
print(missing_values)
```

```python
missing_percent_values = (df.isnull().sum() / len(df)) * 100
print("missing value percent:")
print(missing_percent_values.round(2))
```

```python
columns_missing_values = df.columns[df.isnull().any()].tolist()
print("Columns with missing values:")
print(columns_missing_values)
```

```python
rows_mising_values = df[df.isnull().any(axis=1)]
print("Rows with missing values:")
rows_mising_values
```

```python
plt.figure(figsize = (20, 10))
sns.heatmap(df.isnull())
plt.title("Missing values in the data")
plt.xlabel("Columns")
plt.xticks(range (len(df.columns)), df.columns, rotation = 45)
plt.ylabel("Rows")
plt.show()
```
