```python
import pandas as pd         #import panda for data manipulation
```

```python
data = {
    "Name": ["Alice", "BOB","charlie", "David", "Eve"],        # data making in dataframe
    "Math": [85,65,90,70,95],
    "science": [65,80,75,85,85],
    "English": [90,85,65,59,80],
    "History": [70,75,80,90,95],
    "Grade": ["A", "A+","A","B","B+"]
    }
df = pd.DataFrame(data)
df = pd.DataFrame(data,index=["s1","s2","s3","s4","s5"])     #creating data frame with index
print("Our student marks Table:")
df
```

```python
#---------------lets understand the indexing of our table--------------
print("shape (rows, columns):",df.shape)     #shape of the data frame
print("Row labels(index):",df.index.tolist())      #index of the data frame and name of rows
print("columns labels:", df.columns.tolist())     #columns of dataframe and name of columns
```

```python
df.info()      #information about the dataframe
```

```python
df.describe()     #statistical summary of the dataframe
```

#-------------.Loc- pick by label(Name)
syntax: df.loc[row_label, column_label]
.loc actual name of  uses row and column 

```python
df
```

```python
#-------------example of .Loc------------------
print("Marks of student s1 in all subjects:")
print(df.loc["s1"])   #Loc is used to access the row with index s1 and all columns
```

```python
print("Marks od student s4 in all subjects:")
print(df.loc["s4"]) 
```

```python
df
```

```python
#-----------------One student marks in one subject-----------------
print("Bob's marks:", df.loc["s2"]["Math"])     #marks of student s2(BOB) in math subject
print("Eve's marks:", df.loc["s5"]["History"])     #marks of student s5(Eve) in history subject
print("Charlie's marks:", df.loc["s3"]["English"])     #marks of student s3(charlie) in english subject
print("Alice's grade:", df.loc["s1"]["Grade"])     #grade of student s1(Alice) in grade column
```

```python
#--------------------one student two subjects marks--------------------
print("David's Marks in science and history:")
print(df.loc["s4",["science","History"]])    #marks of student s4(David) in science and history subjects
```

```python

math_BOB = df.loc["s2","Math"] #marks of student s2(BOB) in math subject
print("BOB marks of math:", math_BOB)
```

```python
df
```

```python
#-------------------Print marks of one student of all subjects-------------
print("Marks of student Charlie and david of all subjects:")
print(df.loc[["s3","s4"]])  #marks of student s3(charlie) and s4(david) in all subjects and grade column
```

```python
#----------------------only math and science marks of all students----------------------
print("Name, Math and science marks of all students:")
print(df.loc[:,["Name","Math","science"]])  #math and science marks of all students
```

```python
#----------------only name and grade of two students-----------------
print("Name and grade of student Alice and Eve:")
print(df.loc[["s1","s5"],["Name", "Grade"]]) #name and grade of student s1(Alice) and s5(Eve)
```

part 2: .iloc  pick by position number
.iloc uses numbers(0,1,2,3,4-------)to pick rows and columns

```python
df
```

```python
print(df.iloc[0])    #iloc is used to access the row and column by their integer position, here it will return the value at row 1 and column 3 which is 85
```

```python
print(df.iloc[2,1])    #marks of student s3(charlie) in math subject using iloc
```

```python
print("First three rows of student marks table:")
print(df.iloc[0:3])    #first three rows of student marks table using iloc
```

```python
print("Last one row of student marks table:")
print(df.iloc[-1:-2:-1])    #last one row of student marks table using iloc
```

```python
print("last two rows of student marks table:")
print(df.iloc[-2:])    #last two rows of student marks table using iloc
```

```python
print("From last to first row of student marks table:")
print(df.iloc[::-1])    #from last to first row of student marks table using iloc
```

```python
print("First student marks in all subjects:")
print(df.iloc[0])    #first student marks in all subjects using iloc
```

```python
value = df.iloc[4,4] #row position 4 and column position 4
print("value at row position 4 and column position 4:", value)
```
