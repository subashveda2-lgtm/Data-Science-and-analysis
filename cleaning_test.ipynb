```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_raw = pd.read_csv(r"C:\Users\ssita\pythonclass\nepal_survey.csv")
print("The total number of rows is:", df_raw.shape[0])
print("The total number of columns is:", df_raw.shape[1])
df_raw.head(30)
```

```python
missing_count_value = df_raw.isnull().sum() #   missing values in each column
print("Missing count value for each column:")
print(missing_count_value)
```

```python
missing_percent_value = (df_raw.isnull().sum() /len(df_raw) * 100) # percentage of missing values in each column
print("Missing Percent Value is:")
missing_percent_value
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df_raw.isnull())
plt.title("Missing Values heatmap")
plt.xlabel("columns")
plt.ylabel("rows")
plt.xticks(range(len(df_raw.columns)), df_raw.columns, rotation=45)
plt.show()
```

```python
df_clean = df_raw.copy()
print("The total number of rows is:", df_clean.shape[0])
print("The total number of columns is:", df_clean.shape[1])
```

```python
df_clean = df_clean.drop(columns = ["remarks", "latitude_gps", "survey_notes",])
print("The total number of row is:",df_clean.shape[0])
print("The Total number of columns is:", df_clean.shape[1])
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df_clean.isnull())
plt.title("Missing values heatmap")
plt.xlabel("columns")
plt.ylabel("rows")
plt.xticks(range(len(df_clean.columns)), df_clean.columns, rotation= 45)
plt.show()
```

```python
df_clean.head(10)
```

```python
#------------------mean value in household size------------------------
mean = df_clean["household_size"].mean()
print("the mean value is:", mean)

median = df_clean["household_size"].median()
print("the median value is:",median)
```

```python
#--------------------mean value in household size---------------
print("Missing value in household size is:", df_clean["household_size"].isnull().sum())
df_clean["household_size"] = df_clean["household_size"].fillna(mean)
print("mean average value of household size:", round(mean,2))
print("missing value in household size after filling is:", df_clean["household_size"].isnull().sum())
```

```python
missing_count_percent_value = df_clean.isnull().sum() / len(df_clean) * 100
print("Missing count and percent value for each column:")
missing_count_percent_value
```

```python
print("Missing value of age is:", df_clean["age"].isnull().sum())
median_age = df_clean["age"].median()
print("median value of age is:", median_age)
df_clean["age"] = df_clean["age"].fillna(median_age)
print("missing value of age after filling is:", df_clean["age"].isnull().sum())
```

```python
print("Missing value of monthly income nrs is:", df_clean["monthly_income_nrs"].isnull().sum())
median_monthly = df_clean["monthly_income_nrs"].median()
print("Median value is:", median_monthly)
df_clean["monthly_income_nrs"] = df_clean["monthly_income_nrs"].fillna(median_monthly)
print("missing value of monthly income nrs after filling is:", df_clean["monthly_income_nrs"].isnull().sum())
```

```python
print("Missing vlur of land ownership is:", df_clean["land_ownership_ropani"].isnull().sum())
median_land = df_clean["land_ownership_ropani"].median()
print("Median value is:", median_land)
df_clean["land_ownership_ropani"] = df_clean["land_ownership_ropani"].fillna(median_land)
print("Missing value of land ownership after filling is:", df_clean["land_ownership_ropani"].isnull().sum())
```

```python
print("Missing value of remmitance is:", df_clean["annual_remittance_nrs"].isnull().sum())
median_remittance = df_clean["annual_remittance_nrs"].median()
print("median value is:", median_remittance)
df_clean["annual_remittance_nrs"] = df_clean["annual_remittance_nrs"].fillna(median_remittance)
print("Missing value of remittance after filling is:", df_clean["annual_remittance_nrs"].isnull().sum())
      
```

```python
df_clean["occupation"].value_counts()
```

```python
#-------------------------------put missing value of occupation using mode---------------
mode_occupation = df_clean["occupation"].mode()[0]
print("The mode value of occupation is:", mode_occupation)
print("missing value of occupation is:", df_clean["occupation"].isnull().sum())
df_clean["occupation"] = df_clean["occupation"].fillna(mode_occupation)
print("missing value of occupation after filling is:", df_clean["occupation"].isnull().sum())
```

```python
df_clean["education_level"].value_counts()
```

```python
mode_education = df_clean["education_level"].mode()[0]
print("The mode value of education level is:", mode_education)
print("missing value of education level is:", df_clean["education_level"].isnull().sum())
df_clean["education_level"] = df_clean["education_level"].fillna(mode_education)
print("missing value of education level after filling is:", df_clean["education_level"].isnull().sum())
```

```python
#---------------ffill using on drinking water source--------------
print("Missing value of drinking water source is:", df_clean["drinking_water_source"].isnull().sum())
df_clean["drinking_water_source"] = df_clean["drinking_water_source"].ffill()
print("missing value of drinking water source after filling is:", df_clean["drinking_water_source"].isnull().sum())
```

```python
#----------------------bfill using in health insurance-----------------
print("Missing value of health insurance is:", df_clean["health_insurance"].isnull().sum())
df_clean["health_insurance"] = df_clean["health_insurance"].bfill()
print("missing value of health insurance after filling is:", df_clean["health_insurance"].isnull().sum())
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df_clean.isnull())
plt.title("missing values heatmap")
plt.xlabel("columns")
plt.ylabel("rows")
plt.xticks(range(len(df_clean.columns)), df_clean.columns, rotation=90)
plt.show()
```

```python
missing_value_count_percent = df_clean.isnull().sum() / len(df_clean) * 100
print("Missing count and percent value for each column:")
missing_value_count_percent
```

```python
#----------------------save in csv file------------------
df_clean.to_csv("test_cleaned_data.csv", index=False)
```
