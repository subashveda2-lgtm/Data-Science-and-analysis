```python
import pandas as pd                      #-------to import the pandas library
import matplotlib.pyplot as plt        #-------to import the matplotlib library for data visualization
import seaborn as sns                #-------to import the seaborn library for data visualization
```

```python
df = pd.read_csv(r"C:\Users\ssita\pythonclass\student_ledger.csv") #-------to read the dataset
print("shape of the dataset", df.shape) #-------to get the shape of the dataset
df
```

```python
df.info() #-------to get the information about the dataset
```

```python
missing_count = df.isnull().sum() #-------to get the count of missing values in each column
print("Count of missing values:") #-------to display the count of missing values in each column
print(missing_count) #-------to display the count of missing values in each column
```

```python
missing_percent = (df.isnull().sum() / len(df)) *100 #-------to calculate the percentage of missing values in each column
print("percentage of missing values:") #-------to display the percentage of missing values in each column
print(missing_percent.round(2)) #-------to display the percentage of missing values in each column rounded to 2 decimal places
```

```python
columns_missing_values = df.columns[df.isnull().any()].tolist() #-------to get the list of columns that have missing values
print("columns with missing values:") #-------to display the list of columns that have missing values
print(columns_missing_values) #-------to display the list of columns that have missing values
```

```python
rows_missing_values = df[df.isnull().any(axis=1)] #-------to get the rows that have missing values
print("rows with missing values:", len(rows_missing_values)) #-------to display the rows that have missing values
rows_missing_values #-------to display the rows that have missing values
```

```python
plt.figure(figsize = (15,10)) #-------to set the figure size for the plot
plt.title("Distribution of Grades") #-------to set the title of the plot
plt.imshow(df.isnull(), aspect = "auto", cmap = "Greys", interpolation = "none") #-------to create a heatmap to visualize the missing values in the dataset
plt.xticks(range(len(df.columns)), df.columns, rotation = 45, ha ="right") #-------to set the x-axis ticks and labels
plt.yticks(range(len(df)), df["StudentID"]) #-------to set the y-axis ticks and labels"])
plt.xlabel("columns") #-------to set the x-axis label
plt.ylabel("StudentID") #-------to set the y-axis label
plt.tight_layout() #-------to adjust the layout of the plot
plt.show() #-------to display the plot
```

```python
plt.figure(figsize=(15,8)) #-------to set the figure size for the plot
sns.heatmap(df.isnull()) #-------to create a heatmap to visualize the missing values in the dataset using seaborn library
plt.title("Heatmap of Missing Values") #-------to set the title of the plot
plt.xlabel("columns") #-------to set the x-axis label
plt.xticks(range(len(df.columns)), df.columns, rotation=45, ha="right") #-------to set the x-axis ticks and labels
plt.ylabel("StudentID") #-------to set the y-axis label
plt.show() #-------to display the plot
```

Step 5: dropna() removes rows with missing values.

```python
#----------------Drop rows where all values are missing----------------
df_drop_all = df.dropna(how = "all") #-------to drop rows where all values are missing
print("Rows Before dropping rows:", len(df)) #-------to display the number of rows before dropping rows with all missing values
print("Rows after dropping rows:",len(df_drop_all)) #-------to display the number of rows after dropping rows with all missing values
```

```python
#---drop rows where any value is missing---
df_drop_any = df.dropna(how = "any") #-------to drop rows where any value is missing
print("Rows after dropping rows:",len(df_drop_any)) #-------to display the number of rows after dropping rows with any missing values
```

Step 5 drop the missing value indicates by dropna()

```python
#-------------drop rows only if specific columns have missing values----------------
df_drop_subset = df.dropna(subset = ["FinalGrade", "MidtermGrade", "Section"]) #-------to drop rows only if specific columns have missing values
print("Rows after dropping rows of grade and midterm_grade:", len(df_drop_subset)) #-------to display the number of rows after dropping rows with missing values in specific columns
```

```python
#_-------------------drop columns that have more than 30% missing values-----------------
threshold = len(df) * 0.70 #-------to set the threshold for dropping columns with more than 30% missing values
df_drop_columns = df.dropna(axis = 1, thresh = int(threshold)) #-------to drop columns that have more than 30% missing values
print("Columns after Remaining:", df_drop_columns.columns.tolist()) #-------to display the remaining columns after dropping columns with more than 30% missing values
```

Step 6: Filled with Fixed Value
fillna() replaces missing values with a you provided.

```python
df
```

```python
#------------make a working copy so we do not modify the original---------------------
df_work = df.copy() #-------to create a working copy of the original dataset to avoid modifying it
```

```python
#-------------------fill missing values midterm grade with not available----------------
df_work["MidtermGrade"]= df_work["MidtermGrade"].fillna("Not Available") #-------to fill missing values in the "MidtermGrade" column with "Not Available"
df_work["FinalGrade"] = df_work["FinalGrade"].fillna("Not Available") #-------to fill missing values in the "FinalGrade" column with "Not Available"
df_work["Section"] = df_work["Section"].fillna("A") #-------to fill missing values in the "Section" column with "A"
print("midterm grade after filling:", df_work["MidtermGrade"].isnull().sum()) #-------to display the count of missing values in the "MidtermGrade" column after filling
print("final grade after filling:", df_work["FinalGrade"].isnull().sum()) #-------to display the count of missing values in the "FinalGrade" column after filling
print("section after filling:", df_work["Section"].isnull().sum()) #-------to display the count of missing values in the "Section" column after filling
```

```python
plt.figure(figsize=(20,8)) #-------to set the figure size for the plot
sns.heatmap(df_work.isnull()) #-------to create a heatmap to visualize the missing values in the working copy of the dataset using seaborn library
plt.title("heatmap of missing values") #-------to set the title of the plot
plt.xlabel("columns") #-------to set the x-axis label
plt.xticks(range(len(df_work.columns)), df_work.columns, rotation =45, ha = "right") #-------to set the x-axis ticks and labels
plt.ylabel("StudentID") #-------to set the y-axis label
plt.show() #-------to display the plot
```

```python
missing_counting_work = df_work.isnull().sum() #-------to get the count of missing values in each column of the working copy of the dataset
print("New count missing value") #-------to display the new count of missing values in each column of the working copy of the dataset
print(missing_counting_work) #-------to display the new count of missing values in each column of the working copy of the dataset
```

```python
missing_counting_work_percent = (df_work.isnull().sum() / len(df_work)) *100 #-------to calculate the percentage of missing values in each column of the working copy of the dataset
print("New percentage of missing values") #-------to display the new percentage of missing values in each column of the working copy of the dataset
print(missing_counting_work_percent.round(2)) #-------to display the new percentage of missing values in each column of the working copy of the dataset rounded to 2 decimal places
```

```python
#------------------ffill missing values with forward fill method----------------
df_work["PhoneNumber"] = df_work["PhoneNumber"].ffill() #-------to fill missing values in the "phoneNumber" column using forward fill method]
print("missing values after foward fill:", df_work["PhoneNumber"].isnull().sum()) #-------to display the count of missing values in the "PhoneNumber" column after forward fill
df_work
```

```python
#---------------------bfill missing values with backward fill method----------------
df_work["Email"] = df_work["Email"].bfill() #-------to fill missing values in the "Email" column using backward fill method
print("missing values after backward fill:", df_work["Email"].isnull().sum()) #-------to display the count of missing values in the "Email" column after backward fill
df_work
```
